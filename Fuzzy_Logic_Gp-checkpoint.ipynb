```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from skfuzzy import control as ctrl
```

```python
# Define fuzzy variables (inputs and outputs)
workload = ctrl.Antecedent(np.arange(0, 101, 1), 'workload')  # Input: Workload (0-100)
storage_demand = ctrl.Antecedent(np.arange(0, 101, 1), 'storage_demand')  # Input: Storage Demand (0-100)
ergonomic_risk = ctrl.Consequent(np.arange(0, 101, 1), 'ergonomic_risk')  # Output: Ergonomic Risk (0-100)
```

```python
#Define membership functions for the variables
workload['low'] = fuzz.trimf(workload.universe, [0, 0, 50])
workload['medium'] = fuzz.trimf(workload.universe, [25, 50, 75])
workload['high'] = fuzz.trimf(workload.universe, [50, 100, 100])
```

```python
# Storage Demand
storage_demand['low'] = fuzz.trimf(storage_demand.universe, [0, 0, 50])
storage_demand['medium'] = fuzz.trimf(storage_demand.universe, [25, 50, 75])
storage_demand['high'] = fuzz.trimf(storage_demand.universe, [50, 100, 100])
```

```python
# Ergonomic Risk
ergonomic_risk['low'] = fuzz.trimf(ergonomic_risk.universe, [0, 0, 50])
ergonomic_risk['medium'] = fuzz.trimf(ergonomic_risk.universe, [25, 50, 75])
ergonomic_risk['high'] = fuzz.trimf(ergonomic_risk.universe, [50, 100, 100])
```

```python
#rule define
rule1 = ctrl.Rule(workload['high'] & storage_demand['high'], ergonomic_risk['high'])
rule2 = ctrl.Rule(workload['medium'] & storage_demand['medium'], ergonomic_risk['medium'])
rule3 = ctrl.Rule(workload['low'] & storage_demand['low'], ergonomic_risk['low'])
```

```python
#inference engine
ergonomic_ctrl = ctrl.ControlSystem([rule1, rule2, rule3])
# Simulation Control System
ergonomic_sim = ctrl.ControlSystemSimulation(ergonomic_ctrl)
```

```python
#visualize
workload.view()
storage_demand.view()
ergonomic_risk.view()
```

```python
#test
ergonomic_sim.input['workload'] = 70
ergonomic_sim.input['storage_demand'] = 80
ergonomic_sim.compute()

print(f"Calculated Ergonomic Risk: {ergonomic_sim.output['ergonomic_risk']}")
```
