```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from skfuzzy import control as ctrl
from functools import reduce
from system import FLSInventoryManagement
```

# Constructing

```python
# Define fuzzy variables (inputs and outputs)
inventory = ctrl.Antecedent(np.arange(0, 51, 1), 'Inventory Level')
demand = ctrl.Antecedent(np.arange(0, 4001, 0.01), 'Demand')
days_before_expired = ctrl.Antecedent(np.arange(0, 51, 1), 'Days Before Expired')

reorder_quantity = ctrl.Consequent(np.arange(0, 51, 1), 'Reorder Quantity')
reorder_urgency = ctrl.Consequent(np.arange(0, 101, 0.01), 'Reorder Urgency')
```

```python
#Define membership functions for the variables
# Inventory Level
inventory['Low'] = fuzz.trapmf(inventory.universe, [0,0, 15, 25])
inventory['Medium'] = fuzz.trimf(inventory.universe, [15, 25, 35])
inventory['High'] = fuzz.trapmf(inventory.universe, [25, 35, 50,50])
inventory.view()
```

```python
# Demand 

demand['Low'] = fuzz.trapmf(demand.universe, [0, 0, 1000, 2000])
demand['Medium'] = fuzz.trimf(demand.universe, [1000, 2000, 3000])
demand['High'] = fuzz.trapmf(demand.universe, [2000, 3000, 4000, 4000])

demand.view()
```

```python
# Days Before Expired
days_before_expired['Low'] = fuzz.trapmf(days_before_expired.universe, [0, 0, 5, 15])
days_before_expired['Medium'] = fuzz.trapmf(days_before_expired.universe, [5,15,25,35])
days_before_expired['High'] = fuzz.trapmf(days_before_expired.universe, [25, 35, 50,50])

days_before_expired.view()
```

```python
# Reorder Quantity
reorder_quantity['Low'] = fuzz.trapmf(reorder_quantity.universe, [0,0, 15, 25])
reorder_quantity['Medium'] = fuzz.trimf(reorder_quantity.universe, [15, 25, 35])
reorder_quantity['High'] = fuzz.trapmf(reorder_quantity.universe, [25, 35, 50,50])

reorder_quantity.view()
```

```python
# Reorder Urgency
reorder_urgency['Low'] = fuzz.trapmf(reorder_urgency.universe, [0,0, 25, 35])
reorder_urgency['Medium'] = fuzz.trapmf(reorder_urgency.universe, [25,35,65,75])
reorder_urgency['High'] = fuzz.trapmf(reorder_urgency.universe, [65, 75, 100,100])

reorder_urgency.view()
```

# Rules

```python
#rules achieve
variablesList = [
    # Demand, Inventory, Daysbfexp, reorder_quantity, reorder_urgency
    ["Low", "Low", "High", "Low", "Low"],
    ["Low", "Medium", "Low", "Low", "Medium"],
    ["Low", "High", "High", "Low", "Low"],
    ["Medium", "High", "High", "Low", "Low"],
    ["Medium", "Low", "Medium", "High", "Medium"],
    ["Medium", "Medium", "Medium", "Medium", "Medium"],
    ["High", "Low", "Low", "High", "High"],
    ["High", "Medium", "Low", "Medium", "High"],
    ["High","High","High","Low","Low"]
]

rules = []

for dem, inv, days, quantity, urgency in variablesList:
    labels = dem[0] + inv[0] + days[0] + quantity[0] + urgency[0]
    fuzzyRule = ctrl.Rule( #2
        demand[dem] & inventory[inv] & days_before_expired[days],
        (reorder_quantity[quantity], reorder_urgency[urgency])
    )
    rules.append(fuzzyRule)
```

```python
#rules achieve
rules = [
    ctrl.Rule(demand['Low'] & inventory['Low'] & days_before_expired['High'], 
              (reorder_quantity['Low'], reorder_urgency['Low'])),
    ctrl.Rule(demand['Medium'] & inventory['Medium'] & days_before_expired['Medium'], 
              (reorder_quantity['Medium'], reorder_urgency['Medium'])),
    ctrl.Rule(demand['High'] & inventory['Low'] & days_before_expired['Low'], 
              (reorder_quantity['High'], reorder_urgency['High'])),
    ctrl.Rule(demand['High'] & inventory['High'] & days_before_expired['High'], 
              (reorder_quantity['Low'], reorder_urgency['Low'])),

]
rules.append(ctrl.Rule(
    (demand['Low'] | demand['Medium'] | demand['High']) &
    (inventory['Low'] | inventory['Medium'] | inventory['High']) &
    (days_before_expired['Low'] | days_before_expired['Medium'] | days_before_expired['High']),
    (reorder_quantity['Low'], reorder_urgency['Low'])
))
```

```python
# Engine
system = ctrl.ControlSystem(rules)
simulation = ctrl.ControlSystemSimulation(system)
```

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy.control import Antecedent, Consequent, Rule, ControlSystem, ControlSystemSimulation

# Input variables
demand = Antecedent(np.arange(0, 4001, 0.01), 'demand')
inventory = Antecedent(np.arange(0, 51, 1), 'inventory')
days_before_expiry = Antecedent(np.arange(0, 51, 1), 'days_before_expiry')

# Output variables
quantity = Consequent(np.arange(0, 51, 1), 'quantity')
urgency = Consequent(np.arange(0, 101, 0.01), 'urgency')

# Membership functions for inputs
inventory['L'] = fuzz.trapmf(inventory.universe, [0, 0, 15, 25])
inventory['M'] = fuzz.trimf(inventory.universe, [15, 25, 35])
inventory['H'] = fuzz.trapmf(inventory.universe, [25, 35, 50, 50])
inventory.view()

demand['L'] = fuzz.trapmf(demand.universe, [0, 0, 1000, 1500])
demand['M'] = fuzz.trimf(demand.universe, [1000, 1750, 2500])
demand['H'] = fuzz.trapmf(demand.universe, [1750, 2500, 4000, 4000])
demand.view()

days_before_expiry['L'] = fuzz.trapmf(days_before_expiry.universe, [0, 0, 5, 15])
days_before_expiry['M'] = fuzz.trapmf(days_before_expiry.universe, [5, 15, 25, 35])
days_before_expiry['H'] = fuzz.trapmf(days_before_expiry.universe, [25, 35, 50, 50])
days_before_expiry.view()

# Membership functions for outputs
quantity['L'] = fuzz.trapmf(quantity.universe, [0, 0, 15,25])
quantity['M'] = fuzz.trimf(quantity.universe, [15, 25, 35])
quantity['H'] = fuzz.trapmf(quantity.universe, [25, 35, 50, 50])
quantity['0'] = fuzz.trimf(quantity.universe, [0, 0, 0])  # Fixed zero
quantity.view()

urgency['L'] = fuzz.trapmf(urgency.universe, [0, 0, 25, 35])
urgency['M'] = fuzz.trapmf(urgency.universe, [25, 35, 65, 75])
urgency['H'] = fuzz.trapmf(urgency.universe, [65, 75, 100, 100])
urgency['0'] = fuzz.trimf(urgency.universe, [0, 0, 0])  # Fixed zero
urgency.view()
```

```python
rules = [
    Rule(demand['L'] & inventory['L'] & days_before_expiry['H'], 
         (quantity['L'], urgency['L'])),
    Rule(demand['L'] & inventory['M'] & days_before_expiry['L'], 
         (quantity['L'], urgency['M'])),
    Rule(demand['H'] & inventory['L'] & days_before_expiry['L'], 
         (quantity['H'], urgency['H'])),
    Rule(demand['H'] & inventory['M'] & days_before_expiry['L'], 
         (quantity['M'], urgency['H'])),
    Rule(inventory['H'] & days_before_expiry['H'], 
         (quantity['0'], urgency['0'])),
]

for l in ['L','M','H']:
     rules.append(Rule(demand['M'] & ~inventory[l] & days_before_expiry['M'], 
         (quantity[l], urgency['M'])))
```

```python
control_system = ControlSystem(rules)
simulation = ControlSystemSimulation(control_system)
```

```python
def plotting_subplot(axes, x, functionList, clr, labels, title, kind, symbol, index, aggregated = None, specific_x_val = None):
    # Determine if `axes` is 1D or 2D
    if axes.ndim == 1:  # 1D array of subplots
        ax = axes[index]
    else:  # 2D array of subplots
        rows, cols = axes.shape
        row = index // cols
        col = index % cols
        ax = axes[row, col]

    # Plot on the specific subplot
    for ind in range(len(functionList)):
        ax.plot(x, functionList[ind], color=clr[ind], label=labels[ind])
    
    if aggregated is not None:
        ax.plot(x, aggregated, color='purple', label="Aggregated")
    
    if specific_x_val is not None:
        ax.plot([specific_x_val, specific_x_val], [0,1], color = 'brown', label = "Value")

    ax.set_title(f"{title} ({kind + symbol})", fontsize=10)
    ax.grid(True)
    ax.legend(fontsize=8)
    ax.set_xlabel("x-axis")
    ax.set_ylabel("Membership Degree")

# Example inputs
x = [x_inventory, x_days_before_expiry, x_quantity, x_urgency]
func = [inventory, days, quantity, urgency]
clrs = [['r','g','b'],['r','g','b'],['k','r','g','b'], ['k','r','g','b']]
labels = [['Low','Medium','High'],['Low','Medium','High'],['Zero','Low','Medium','High'],['Zero','Low','Medium','High']]
title = ["Inventory","Days Before Expired","Reorder Quantity", "Reorder Urgency"]
kind = ['Input','Input','Output','Output']
symbol = [' Units', ' Days', ' Units', ' %']

# Create subplots
rows = 2  # Number of rows in the grid
cols = 2  # Number of columns in the grid
fig, axes = plt.subplots(rows, cols, figsize=(15, 6))
axes = axes.flatten()  # Flatten axes for easier access

# Loop through and plot on subplots
for i in range(len(x)):
    plotting_subplot(axes, x[i], func[i], clrs[i], labels[i], title[i], kind[i], symbol[i], i)

# Remove any extra subplots (if len(x) < rows * cols)
for j in range(len(x), rows * cols):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
def interpert_membership_inputs(x,funcList,val):
    result = {
        "Low": 0,
        "Medium": 0,
        "High": 0
    }
    linguistics = ['Low', 'Medium', 'High']
    for ind in range(len(funcList)):
        result[linguistics[ind]] = fuzz.interp_membership(x, funcList[ind], val)
    return result

def interpert_membership_outputs(x,funcList,val):
    result = {
        "Zero": 0,
        "Low": 0,
        "Medium": 0,
        "High": 0
    }
    linguistics = ['Zero','Low', 'Medium', 'High']
    for ind in range(len(funcList)):
        result[linguistics[ind]] = fuzz.interp_membership(x, funcList[ind], val)
    return result

def fuzzification(inv: int, days_val: int):
    # capped_demand = input_demand(dem)
    capped_inv = input_inventory(inv)
    capped_day = input_days(days_val)

    # Use the correct membership functions
    # fuzzyDemand = interpert_membership_inputs(x_demand, demand, capped_demand)
    fuzzyInv = interpert_membership_inputs(x_inventory, inventory, capped_inv)
    fuzzyDays = interpert_membership_inputs(x_days_before_expiry, days, capped_day)

    return fuzzyInv, fuzzyDays #, fuzzyDemand

invFuzz, daysFuzz = fuzzification(10, 35)
print(invFuzz)
print(daysFuzz)
```

# Inference Engine

```python
rule_list = [
    # Inventory, Days - Reorder_q, Reorder_urgency
    [['Low','Low'],['High','High']],
    [['Medium', 'Low'],['Medium', 'High']],
    [['High', 'Low'],['Medium','High']],
    [['Low', 'Medium'],['High','Medium']],
    [['Medium','Medium'],['Medium','Medium']],
    [['High','Medium'],['Low','Low']],
    [['Low','High'],['Medium','Low']],
    [['Medium','High'],['Low','Zero']],
    [['High','High'],['Zero','Zero']]
]

def intialize_quantity(outputQuantity):
    if outputQuantity == 'Low':
        quantity_set = quantity[1]
    elif outputQuantity == 'Medium':
        quantity_set = quantity[2]
    elif outputQuantity == 'High':
        quantity_set = quantity[3]
    else:
        quantity_set = quantity[0]
    return quantity_set

def intialize_urgency(outputUrgent):
    if outputUrgent == 'Low':
        urgency_set = urgency[1]
    elif outputUrgent == 'Medium':
        urgency_set = urgency[2]
    elif outputUrgent == 'High':
        urgency_set = urgency[3]
    else:
        urgency_set = urgency[0]
    return urgency_set

def inference(invFuzz, daysFuzz, inputsVar, outputsVar, ruleNo):
    inputsInv = inputsVar[0]
    inputsDay = inputsVar[1]
    outputQuantity = intialize_quantity(outputsVar[0])
    outputUrgent = intialize_urgency(outputsVar[1])

    rule = np.fmin(invFuzz[inputsInv], daysFuzz[inputsDay])
    quantity_res = np.fmin(rule, outputQuantity)
    urgency_res = np.fmin(rule, outputUrgent)

    return ruleNo, quantity_res, urgency_res
```

```python
quantity_agg = []
urgency_agg = []

for ind in range(len(rule_list)):
    inputsVar = rule_list[ind][0]
    outputsVar = rule_list[ind][1]
    ruleNo, quantity_res, urgency_res = inference(invFuzz,daysFuzz,inputsVar, outputsVar, ind)
    quantity_agg.append(quantity_res)
    urgency_agg.append(urgency_res)
```

```python
# Aggregate all quantity rules
aggregated_quantity = reduce(np.fmax, quantity_agg)

# Aggregate all urgency rules
aggregated_urgency = reduce(np.fmax, urgency_agg)

print("Aggregated Quantity Membership Function:")
print(aggregated_quantity)
print("\nAggregated Urgency Membership Function:")
print(aggregated_urgency)

quantity_result = fuzz.defuzz(x_quantity, aggregated_quantity, 'centroid')
urgency_result = fuzz.defuzz(x_urgency, aggregated_urgency, 'centroid')

print("Quantity:", quantity_result)
print("Urgency:", urgency_result)

```
